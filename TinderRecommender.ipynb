# Completeness Rating Calculation

```python
def completeness_rating(data_ser, completeness_weights):
    data_np = (data_ser.to_numpy() != 'none') * 1
    completeness_rating = np.dot(data_np, completeness_weights)/np.sum(completeness_weights)
    return completeness_rating
```

# Similarity Rating

```python
def similarity_rating(
        data_user : pd.Series,
        data_target : pd.Series,
):
    return (np.sum(data_user == data_target))
```

# Content Based Filtering Model

```python
class ContentBasedFiltering:
    def __init__(self,num_outputs):
        self.num_outputs = num_outputs
        self.user_NN = None
        self.item_NN = None
        self.model = None

    def user_net(self):
        tf.random.set_seed(1)
        user_NN = tf.keras.models.Sequential([
            tf.keras.layers.Dense(units = self.num_outputs*5, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs*6, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs*7, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs*6, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs*5, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs, activation = 'linear'),
        ])
        self.user_NN = user_NN
        return user_NN
    def item_net(self):
        item_NN = tf.keras.models.Sequential([
            tf.keras.layers.Dense(units = self.num_outputs*5, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs*6, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs*7, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs*6, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs*5, activation = 'relu'),
            tf.keras.layers.Dense(units = self.num_outputs, activation = 'linear'),
        ])
        self.item_NN = item_NN
        return item_NN
    def _model(self):
        # create the user input and point to the base network
        input_user = tf.keras.layers.Input(shape=(self.num_outputs))
        vu = self.user_NN(input_user)
        vu = tf.linalg.l2_normalize(vu, axis=1)

        input_item = tf.keras.layers.Input(shape=(self.num_outputs))
        vm = self.item_NN(input_item)
        vm = tf.linalg.l2_normalize(vm, axis=1)

        output = tf.keras.layers.Dot(axes=1)([vu, vm])
        model = tf.keras.Model([input_user, input_item], output)
        return model;

    def build_model(self):
        self.user_net()
        self.item_net()
        self.model = self._model()
        self._compile();
        return self.model

    def _compile(self):
        cost_fn = tf.keras.losses.MeanSquaredError()
        opt = tf.keras.optimizers.Adam(learning_rate=0.001)
        self.model.compile(optimizer=opt,
                    loss=cost_fn)
```
